```python
# There seems to be no link between engine warranty and price. Let's begin with using Gradient Boosting Regression to predict the 
# bike prices.

# One-hot encode the categorical columns. Convert categorical variables to dummy variables
df_encoded = pd.get_dummies(df, columns=['Bike_company', 'Bike_model', 'Engine_type', 'Fuel_type', "CC(Cubic capacity)", 'Fuel_Capacity'])

# Split data into features and target
X = df_encoded.drop(['Price', 'S.no'], axis=1)
y = df_encoded['Price']
```

```python
# Assuming X contains the features and y contains the target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Creating Random Forest Regression model
rf_model = RandomForestRegressor(n_estimators=36, max_depth=10, random_state=0)
```

```python
# Training the model on the training set
rf_model.fit(X_train, y_train)
```

```python
# Predicting on the test data
y_pred = rf_model.predict(X_test)
```

```python
# setting the number of folds for k-fold cross-validation
k_folds = KFold(n_splits=20, shuffle=True, random_state=42)
```

```python
# evaluating the model using k-fold cross-validation
mae_scores = -cross_val_score(rf_model, X, y, cv=k_folds, scoring="neg_mean_absolute_error")
mse_scores = -cross_val_score(rf_model, X, y, cv=k_folds, scoring="neg_mean_squared_error")
rmse_scores = np.sqrt(mse_scores)
```

```python
# Printing the average performance metrics
print("Mean Absolute Error:", mae_scores.mean())
print("Mean Squared Error:", mse_scores.mean())
print("Root Mean Squared Error:", rmse_scores.mean())
```
